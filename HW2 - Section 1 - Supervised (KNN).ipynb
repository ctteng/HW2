# Function Definitions

```python
#taken from tutorial 
# at --> http://www.astro.washington.edu/users/vanderplas/Astr599/notebooks/18_IntermediateSklearn
def plot_confusion_matrix(y_pred, y):
    plt.imshow(metrics.confusion_matrix(y, y_pred),
               cmap=plt.cm.binary, interpolation='nearest')
    plt.colorbar()
    plt.xlabel('true value')
    plt.ylabel('predicted value')
```

```python
def gen_confusion_matrix(x_tst, x_trn, y_tst, y_trn):
    knn_classifier = neighbors.KNeighborsClassifier(n_neighbors=1)
    knn_classifier.fit(x_trn, y_trn)
    y_predictor = knn_classifier.predict(x_tst)
    print "classification accuracy:", metrics.accuracy_score(y_tst, y_predictor)
    print x_trn.shape, x_tst.shape
    plot_confusion_matrix(y_tst, y_predictor)
```

```python
#taken from GA HW 6
def plot_knn_scores(x_trn, x_tst, y_trn, y_tst, wine_neighbors):
    # Loop through each neighbors value from 1 to 51 and append
    # the scores
    scores = []
    
    for n in wine_neighbors:
        clf = neighbors.KNeighborsClassifier(n)
        clf.fit(x_trn, y_trn)
        scores.append(clf.score(x_tst, y_tst))
        
    plt.figure(figsize=(18,6))
    _ = plt.plot(wine_neighbors, scores, linewidth=3.0)
    
    knn_df = pd.DataFrame(zip(wine_neighbors, scores))
    knn_df.columns =['k-count','knn-score']
    knn_df.set_index(['knn-score'])
    return knn_df
```

```python
def plot_bar_chart(column_names, x_data, y_data):
    n = np.arange(len(column_names))
    fig = plt.figure(figsize=(18,6))
    ax = fig.add_subplot(111)
    ax.bar(n, feature_selection.f_classif(x_data, y_data)[0])
    xtickNames = ax.set_xticklabels(column_names)
    ax.set_xticks(n)
    _ = plt.setp(xtickNames, rotation=45, fontsize=16)
    
```

```python
def cross_validation(x_data,target_data, trial_count):
    clf = neighbors.KNeighborsClassifier(11, weights='uniform')
    scores = cross_val_score(clf, x_data, target_data, cv=trial_count)
    print 'Mean of CV scores:: ' + str(np.mean(scores))
    print 'Standard deviation of CV scores::' + str(np.std(scores))
    print 'Cross Validation scores:: ' + np.array_str(scores)
```

```python
def get_massage_data(path,do_normalize):
    wine_df = pd.read_csv(path)
    wine_params_df = wine_df.drop('Wine', axis=1)
    wine_target_df = wine_df['Wine']
    wine_data_nump = wine_params_df.values
    wine_target_nump = wine_target_df.values
    
    if(do_normalize == True):
        wine_data_scaled = scale(wine_data_nump)
        return wine_data_scaled, wine_target_nump, wine_params_df.columns.values
    else: 
        return wine_data_nump, wine_target_nump, wine_params_df.columns.values
```

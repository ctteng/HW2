# Data Prep and Package Imports

```python
%matplotlib inline

import numpy as np
import pandas as pd

from seaborn import plt, color_palette
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
from __future__ import division
```

```python
wine_df = pd.read_csv("C:\\Development\\DAT_SF_12\\DAT_SF_12\\data\\Wine.csv")
```

```python
wine_ndarray = wine_df.drop('Wine', axis=1).values
wine_actual_labels = wine_df['Wine'].values
wine_scaled = scale(wine_ndarray)
```

```python
wine_name_list = list(wine_df.drop('Wine', axis=1).columns)
```

# KMeans Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(wine_ndarray)
```

```python
kmeans.labels_
```

```python
wine_actual_labels
```

```python
kmeans_compare_df = pd.DataFrame({ 'actual_lables' : wine_actual_labels, 
                                  'kmeans_labels': kmeans.labels_, 
                                  'reclassified_kmeans_lables' : kmeans.labels_,
                                  'is_match' : pd.Series(1,index=list(range(len(wine_actual_labels))), dtype='bool')})
#relabel for comparison
#i know there is a better way to do this...will have to come back to this later
kmeans_compare_df.loc[kmeans_compare_df['kmeans_labels'] == 1, 'reclassified_kmeans_lables'] = 1
kmeans_compare_df.loc[kmeans_compare_df['kmeans_labels'] == 0, 'reclassified_kmeans_lables'] = 2
kmeans_compare_df.loc[kmeans_compare_df['kmeans_labels'] == 2, 'reclassified_kmeans_lables'] = 3
#set field is_match to True/False for counting accurate labels
kmeans_compare_df.loc[kmeans_compare_df['actual_lables'] != kmeans_compare_df['reclassified_kmeans_lables'], 'is_match'] = False
#get percentage of accurate labels
print 'Accuracy rate:::' +  str(len(kmeans_compare_df[ kmeans_compare_df['is_match'] == True])/len(kmeans_compare_df))
```

# Hierarchical Clustering

```python
distx = squareform(pdist(wine_scaled, metric='euclidean'))
dist_hamming = squareform(pdist(wine_scaled, metric='hamming'))
```

```python
R = dendrogram(linkage(distx, method='single'))

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
R = dendrogram(linkage(dist_hamming, method='single'))

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
R = dendrogram(linkage(distx, method='ward'))

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
R = dendrogram(linkage(dist_hamming, method='ward'))

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```
